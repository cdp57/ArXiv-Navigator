```python
import pandas as pd
df=pd.read_csv('Arxiv_Normalized.csv')
```

```python

topics = {}

for k in range(1, 4):
    for i, j in enumerate(df['Category ' + str(k)]):
        if pd.isna(j) == False:
            if j in topics:
                topics[j] += 1
            else:
                topics[j] = 1

hold = list(topics.items())

aut = dict(sorted(topics.items(), key=lambda item: item[1], reverse=True))

print(aut)

saved = list(aut.items())

ts = []
nums = []

for i in range(0,10):
    ts.append(saved[i][0])
    nums.append(saved[i][1])

name = {'a':ts, 'b': nums}

dataframe = pd.DataFrame(name)

print(dataframe)

ax = sns.barplot(dataframe, x='a', y='b', color='crimson')

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.title('Arxiv\'s STEM Content', fontdict={'size':22})
plt.xlabel('Most Common Arxiv Categories', fontdict={'size':13})
plt.ylabel('Number of Articles', fontdict={'size':13})
plt.show()
```

```python
import matplotlib.pyplot as pyplot

from lmfit import Model

x2_data = EC
time = NC



def model(b, x, y, time):
    
    return (b  +  time**x+y)


gmodel = Model(model, independent_vars=['time'])
result = gmodel.fit(x2_data, time=time, b=1, x=-0, y=-0)


print(result.fit_report())


plt.axhline(y=9, color = 'r', linestyle = 'dashed')
plt.legend(['EPA Safe Levels for FM-2.5'])
plt.plot(time, x2_data, 'o')
# plt.plot(time, result.init_fit, '--', label='initial fit')
plt.plot(time, result.best_fit, '-', label='best fit')



plt.xlabel('Years since 1990')
plt.ylabel('FM-2.5 Levels (Micrograms/cubic meter)')
plt.title('Cleveland FM-2.5 Levels over Time')
```
